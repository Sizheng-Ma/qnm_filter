```python
import matplotlib.pyplot as pl
import numpy as np
```

```python
data = np.loadtxt('./results/221_NNI_random_mass_amp_phase_ALL.dat')
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
sort_idx = np.argsort(SNRtot)
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data[sort_idx].T
```

```python
pl.plot(SNRtot**2, evidence220_221 - nofilter, 'o')
pl.axhline(5, c='r', ls='--')
# pl.xlim(0, 50)
# pl.ylim(0, 20)
```

```python
slope_arr = []
for i in range(len(SNR221)//5):
    slope, _ = np.polyfit((SNR221**2)[5*i:5*i+5], (evidence220_221 - evidence220)[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx((SNR221**2)[::5], slope_arr, 'o')
print(slope_arr[-1])
```

```python
data = np.loadtxt('./results/221_NNI_random_mass_amp_ALL.dat')
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
pl.plot(SNR221**2, evidence220_221-evidence220, 'o', label='Randomised phase')

data = np.loadtxt('./results/221_NNI_random_chi_ALL.dat')
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
pl.plot(SNR221**2, evidence220_221-evidence220, 'o', label='Randomised chi')
pl.xlabel("(221 SNR)^2")
pl.ylabel("ln(BF) 220+221:220")
pl.legend()
```

```python
data = np.loadtxt('./results/221_NNI_random_chi_ALL.dat')
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
pl.plot(SNR221**2, evidence220_221-evidence220, 'o', label='Randomised chi')
pl.xlabel("(221 SNR)^2")
pl.ylabel("ln(BF) 220+221:220")
pl.legend()
pl.xlim(-5, 1600)
```
