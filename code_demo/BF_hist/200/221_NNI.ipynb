```python
import matplotlib.pyplot as pl
import numpy as np
```

```python
data = np.loadtxt('./results/220+221_NNI.dat')
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data.T
sort_idx = np.argsort(SNRtot)
[SNRtot, SNR220, SNR221, evidence220_221, evidence220, evidence221, nofilter] = data[sort_idx].T
```

```python
#Slope of nofilter

slope_arr = []
sort_idx = np.argsort(SNRtot)
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit((SNRtot**2)[5*i:5*i+5], nofilter[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(SNRtot[::5], slope_arr)

pl.title("Slope of 5 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```

```python
#Slope of BF(2:0 modes)

slope_arr = []
sort_idx = np.argsort(SNRtot)
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit((SNRtot**2)[5*i:5*i+5], (evidence220_221 - nofilter)[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx((SNRtot**2)[::5], slope_arr)

pl.title("Slope of 5 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```

```python
#Slope of BF(2:1 modes)

slope_arr = []
sort_idx = np.argsort(SNR221)
for i in range(len(SNR221)//5):
    slope, _ = np.polyfit((SNR221**2)[5*i:5*i+5], (evidence220_221 - evidence220)[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx((SNR221**2)[::5], slope_arr)

pl.title("Slope of 5 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```
