```python
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.offsetbox import AnchoredText
```

## Bayes factor always prefers the correct model, but is SNR dependent

```python
result = np.loadtxt('./results/OF_fixedAmp_results.dat')
reshaped = np.reshape(result, (len(result)//10, 10))
[snr, nofilter, evidence220, evidence221, evidence220_221, evidence220_221_320, occams220_95, \
 occams221_95, occams220_221_95, occams220_221_320_95] = reshaped.T
```

```python
idx = snr>3
pl.plot(snr[idx], (evidence220_221_320[idx] - evidence220[idx]), 'o')
pl.xlabel("SNR")
pl.ylabel("BF(220+221+320 / 220)")
pl.title("3 modes injected")
```

```python
temp0 = ((evidence220_221_320 - nofilter))/snr**2
temp1 = ((evidence220_221 - nofilter))/snr**2
temp2 = ((evidence220 - nofilter))/snr**2
pl.plot(snr, temp0, 'o', label = 'Filtered 3 modes')
pl.plot(snr, temp1, 'o', label = 'Filtered 2 modes')
pl.plot(snr, temp2, 'o', label = 'Filtered 1 mode')
pl.xlabel(r'$\rho$')
pl.ylabel("'Normalised BF'")
pl.title("Injected 3 modes")
pl.legend()
```
