```python
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=0.5_ALL.dat').T
pl.loglog(SNRtot, evidence220_221-nofilter, 'o', c='tab:blue')
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=2_ALL.dat').T
pl.loglog(SNRtot, evidence220_221-nofilter, 'o', c='tab:blue')
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=0.25_ALL.dat').T
pl.loglog(SNRtot, evidence220_221-nofilter, 'o', c='tab:blue')
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=infty_ALL.dat').T
pl.loglog(SNRtot, evidence220_221-nofilter, 'o', c='tab:blue')
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/changed_acf_ratio=0.25_ALL.dat').T
pl.loglog(SNRtot, evidence220_221-nofilter, 'o', c='tab:blue')
pl.title("Injected various 220+221 waveforms with different amplitude ratios")
pl.xlabel("Total SNR")
pl.ylabel("ln(BF) of 220+221:nofilter")
```

```python
[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=0.25_ALL.dat').T
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10((evidence221-nofilter)[sort_idx])
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit(log_snr[5*i:5*i+5], log_BF[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::5], slope_arr, label='Ratio = 0.25')

[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=0.5_ALL.dat').T
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10((evidence221-nofilter)[sort_idx])
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit(log_snr[5*i:5*i+5], log_BF[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::5], slope_arr, label='Ratio = 0.5')

[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=1_ALL.dat').T
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10((evidence221-nofilter)[sort_idx])
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit(log_snr[5*i:5*i+5], log_BF[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::5], slope_arr, label='Ratio = 1')

[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=2_ALL.dat').T
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10((evidence221-nofilter)[sort_idx])
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit(log_snr[5*i:5*i+5], log_BF[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::5], slope_arr, label='Ratio = 2')

[SNRtot, SNR221, evidence220, evidence221, evidence220_221, nofilter] = np.loadtxt('./results/ratio=infty_ALL.dat').T
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10((evidence221-nofilter)[sort_idx])
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit(log_snr[5*i:5*i+5], log_BF[5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::5], slope_arr, label=r'Ratio = $\infty$')

pl.title("Slope of 5 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```
