```python
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.offsetbox import AnchoredText
```

## NNI

```python
result = np.loadtxt('./results/330_amp_ratio=1_ALL.dat')
[SNRtot, SNR330, SNR220, evidence220, evidence330, evidence220_330, nofilter] = result.T
sort_idx = np.argsort(SNRtot)
[SNRtot, SNR330, SNR220, evidence220, evidence330, evidence220_330, nofilter] = result[sort_idx].T
```

## nofilter

```python
pl.plot(SNRtot**2, evidence220_330 - nofilter, 'o', c='tab:blue')
```

```python
slope_arr = []
sort_idx = np.argsort(SNRtot)
for i in range(len(SNRtot)//5):
    slope, _ = np.polyfit((SNRtot**2)[sort_idx][5*i:5*i+5], (evidence220_330 - nofilter)[sort_idx][5*i:5*i+5], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(SNRtot[sort_idx][::5], slope_arr)

pl.title("Slope of 5 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```
