```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
data = np.loadtxt('./results/collated.dat')
```

```python
[SNRtot, credibility] = data.T
```

```python
pl.loglog(SNRtot, credibility, 'o')
pl.xlabel("Total SNR (MF)")
pl.ylabel("ln(BF): 220+221+320:nofilter")
```

```python
slope_arr = []
sort_idx = np.argsort(SNRtot)
log_snr = np.log10(SNRtot[sort_idx])
log_BF = np.log10(credibility[sort_idx])
for i in range(len(SNRtot)//20):
    slope, _ = np.polyfit(log_snr[20*i:20*i+20], log_BF[20*i:20*i+20], deg=1, cov=False)
    slope_arr.append(slope)
pl.semilogx(10**log_snr[::20], slope_arr)

pl.title("Slope of 20 consecutive points at different SNRs")
pl.xlabel("Total SNR")
pl.ylabel("Slope")
pl.legend()
```
