```python
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.offsetbox import AnchoredText
```

## 1 mode injected

```python
result = np.loadtxt('./results/confusion_matrix/1modes_injected_ALL.dat')
reshaped = np.reshape(result, (len(result)//4, 4))
[snr, nofilter, evidence220, evidence220_221] = reshaped.T
```

```python
fit220 = np.polyfit(np.log10(snr), np.log10(evidence220 - nofilter), deg=1,cov=True)
slope220 = fit220[0][0]
uncert220 = fit220[1][0,0]

fig, ax = pl.subplots(1,1)
ax.plot(snr, evidence220-nofilter, 'o')
anchored_text = AnchoredText(r"Slope=%.2f $\pm$ %.2f" % (slope220, uncert220), loc=2)
ax.add_artist(anchored_text)
ax.set_xlabel('SNR')
ax.set_ylabel('ln(BF)')
```

### Low SNR

```python
result = np.loadtxt('./results/overfiltered_1mode_lowSNR_ALL.dat')
reshaped = np.reshape(result, (len(result)//6, 6))
[snr, nofilter, evidence220, evidence220_221, evidence220_210, evidence220_200] = reshaped.T
```

```python
# without normalising for SNR
concat = np.array([evidence220_221-evidence220, evidence220_210-evidence220, evidence220_200-evidence220]).flatten()

pl.plot(snr, evidence220_221 - evidence220, 'o', label = '220+221 vs 220')
pl.plot(snr, evidence220_210 - evidence220, 'o', label = '220+210 vs 220')
pl.plot(snr, evidence220_200 - evidence220, 'o', label = '220+200 vs 220')
pl.axhline(np.median(concat)+2*np.std(concat), c='r', ls='--', label='2 sigma threshold')
pl.legend()
pl.xlabel(r'$\rho$')
pl.ylabel("lg(BF)")
```

```python
result = np.loadtxt('./results/overfiltered_2modes_ALL.txt')
reshaped = np.reshape(result, (len(result)//8, 8))
[snr, nofilter, evidence220, evidence220_221, evidence220_210, evidence220_221_222,
                             evidence220_221_210, evidence220_210_200] = reshaped.T
```

```python
BF_overfiltered222 = evidence220_221_222 - evidence220_221
BF_overfiltered210 = evidence220_221_210 - evidence220_221

pl.plot(snr, BF_overfiltered222, 'o', label='Overfiltered with 222 mode')
pl.axhline(np.median(BF_overfiltered222)+2*np.std(BF_overfiltered222), c='b', ls='--', label='2 sigma for 222')
pl.plot(snr, BF_overfiltered210, 'o', label='Overfiltered with 210 mode')
pl.axhline(np.median(BF_overfiltered210)+2*np.std(BF_overfiltered210), c='orange', ls='--', label='2 sigma for 210')
pl.legend()
```
